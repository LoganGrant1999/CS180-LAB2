```python
import pandas as pd
cereal = pd.read_csv('https://raw.githubusercontent.com/porterjenkins/cs180-intro-data-science/master/data/cereal.csv')
```

## Exercise 1: Protein Powder
Carbs, fats and proteins are the three primary macro nutrients. Create a figure plotting the distribution of each of these macro nutrients together (i.e., three distributions on a single plot). Make sure to provide a legend.

```python
# Enter all of your code for exercise 1 here. Feel free to add more cells if you need to:

import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np


macro_nutrients = cereal[["carbo", "fat", "protein"]]

macros = macro_nutrients.melt(var_name="Nutrient", value_name="Grams")


sns.kdeplot(

    data=macros,

    x="Grams",

    hue="Nutrient",

    common_norm=False,

    fill=False
)

plt.title("Distribution of Macro Nutrients in Cereals")

plt.xlabel("Grams per Serving")

plt.ylabel("Density")

plt.show()
```

## Exercise 2: Sugar Daddy
Get a list of the top 5 most sugary cereals and the 5 least sugary cereals.

```python
# Enter all of your code for exercise 2 here. Feel free to add more cells if you need to:

cereal_sugars = cereal.sort_values("sugars")

least_sugary = cereal_sugars.head(5)[["name", "sugars"]].values.tolist()

most_sugary  = cereal_sugars.tail(5)[["name", "sugars"]].values.tolist()

print("Least Sugary Cereals:")

print(least_sugary)

print("\n")

print("Most Sugary Cereals:")

print(most_sugary)


```

## Exercise 3: Cereal Killer
Get a list of the top 5 highest rated and lowest rated cereals.

```python
# Enter all of your code for exercise 3 here. Feel free to add more cells if you need to:

cereal_ratings = cereal.sort_values("rating")

lowest_rated = cereal_ratings.head(5)[["name", "rating"]].values.tolist()

highest_rated = cereal_ratings.tail(5)[["name", "rating"]].values.tolist()

print("Lowest rated cereals:")

print(lowest_rated)

print("\n")

print("Highest rated cereals:")

print(highest_rated)

```

## Exercise 4: America
Quantify the relationship between sugar and ratings.

Make a plot to visualize this relationship. Superimpose a best fit line (with seaborn) to describe the relationship. It may be helpful to look at the [seaborn regplot documentation]("https://seaborn.pydata.org/generated/seaborn.regplot.html").

```python
# Make the plot for the data visualization and line of best fit here:

sns.regplot(

    data=cereal,

    x="sugars",

    y="rating",
)

plt.title("Relationship between Cereal Sugar and Ratings")

plt.xlabel("Sugar (grams)")

plt.ylabel("Rating")

plt.show()


```


Calculate a correlation statistic describing the relationship between sugar and ratings  (i.e., r or r squared).

```python
# Calculate the statistic using this cell:

r = cereal["sugars"].corr(cereal["rating"])

print("r =", r)
```

Write a statement in plain English interpreting this statistic.

An r = -0.76 indicates that there is a strong negative correlation between sugar content in cereals and cereal rating. Sugary cereals generally have lower ratings

Make five plots comparing the relationships of carbo, sugars, calories, protein, and fat with rating.

```python
# Write your code to compare the various variables with rating below:

sns.regplot(data=cereal, x="carbo", y="rating")

plt.title("Rating vs Carbo")

plt.xlabel("Carbo (grams)")

plt.ylabel("Rating")

plt.show()



sns.regplot(data=cereal, x="sugars", y="rating")

plt.title("Rating vs Sugars")

plt.xlabel("Sugar (grams)")

plt.ylabel("Rating")

plt.show()


sns.regplot(data=cereal, x="calories", y="rating")

plt.title("Rating vs Calories")

plt.xlabel("Calories")

plt.ylabel("Rating")

plt.show()


sns.regplot(data=cereal, x="protein", y="rating")

plt.title("Rating vs Protein")

plt.xlabel("Protein (grams)")

plt.ylabel("Rating")

plt.show()


sns.regplot(data=cereal, x="fat", y="rating")

plt.title("Rating vs Fat")

plt.xlabel("Fat (grams)")

plt.ylabel("Rating")

plt.show()




r_carbo = cereal["carbo"].corr(cereal["rating"])
print("carb r:", r_carbo)

r_sugars = cereal["sugars"].corr(cereal["rating"])
print("sugar r:", r_sugars)

r_calories = cereal["calories"].corr(cereal["rating"])
print("calories r:", r_calories)

r_protein = cereal["protein"].corr(cereal["rating"])
print("protein r:", r_protein)

r_fat = cereal["fat"].corr(cereal["rating"])
print("fat: r", r_fat)
```

Sugar has the strongest relationship with rating. I was able to determine this by calculating the r value for each relationship. It's the closest to 1 or -1 of all the distributions.

## Exercise 7: It’s Hot and It's Cold
The type column has two values: H='hot' and C='cold'. What is the average rating of each type?

```python
# Write your code to get the average rating by type of cereal below:

type_rating = cereal.groupby("type")["rating"].mean()

print(type_rating)
```
